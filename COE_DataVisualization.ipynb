```python
import matplotlib.pyplot as plt
```



>Univariate Analysis : analysis of a single variable to understand its distribution, central tendency, and spread


> Bivariate Analysis : analysis of teo variables tp explore their relationship
    - Common visualization: scatter, line, heatmap

```python
x=[1,2,3,4]
y=[5,6,7,8]
plt.plot(x,y)
plt.show()
```

```python
x=[1,2,3,4]
y=[5,6,7,8]
plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
x=[1,2,3,4]
y=[5,6,7,8]
plt.plot(x,y,color='red',linestyle='dashed',marker='o')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```



> Task : create a line plot showing the growth of a savings account over 6 months with the following :

> => x-axis : months(1-6)


> => y-axis : sasvings account


> Customization : a green dashed line, markers at each data point, appropriate labels and a title


```python
x=[1,2,3,4,5,6]
y=[50000,6000,700000,8000,40000,70000]
plt.plot(x,y,color='green',linestyle='dashed',marker='o')
plt.xlabel('Months')
plt.ylabel('Savings Account')
plt.title('Growth of Savings Account Over 6 Months')
plt.show()
```

```python
categories = ['A','B','C','D','E','F','G']
values = [10,15,7,12,9,30,20]
plt.bar(categories,values,color=['violet','indigo','blue','green','yellow','orange','red'])
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart Example')
plt.show()
```

```python
age=[20,22,25,27,30,35,40,45,50]
b=[20,30,40,50,60,70]
plt.hist(age,bins=b,color='skyblue',edgecolor = "black")
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram Example')
plt.grid(True)
plt.show()
```

```python
programming_lang = ['C','Java','Python','SQL','R']
usage_percentage = [30,25,20,15,10]
plt.pie(usage_percentage,labels=programming_lang,colors=['#FFD1DC','#AEC6CF','#77DD77','#FDFD96','#C3B1E1'],autopct='%1.1f%%')
plt.title('Pie Chart Example')
plt.show()
```

```python
x=[1,2,3,4]
y=[8,3,5,6]
plt.scatter(x,y,color=['#FFD1DC','#AEC6CF','#77DD77','#FDFD96'],s=400,edgecolors='black')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```
