```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```



> # Visualization plots in seaborn
- Histogram (shows distribution of single variable) : sns.histplot()
- KDE Plot (Kernel density Estimation - visualizes the probability density of a variable) : sns.kdeplot()
* Bar plot (display the mean of a qunatative variable for different categories) : sns.barplot()
* Box plot ( used to visualize the distribution of a variable and detect outliers) : sns.boxplot()
* Scatter Plot : sns.scatterplot()
- Line Plot : sns.lineplot()
- Pair Plot : sns.pairplot()
- Heatmap : sns.heatmap()



```python
# histogram plot in seaborn

sns.histplot(data=tips,x='total_bill',bins=20,color='skyblue')
plt.title('Histogram of Total Bill')
plt.show()
```

```python
# histogram plot with kde in seaborn

sns.histplot(data=tips,x='total_bill',bins=10,color='black',kde=True)
plt.title('Histogram of Total Bill')
plt.show()
```

```python
# KDE plot in seborn

sns.kdeplot(data=tips,x='total_bill',color='purple',shade=True)
plt.title('KDE Plot of Total Bill')
plt.show()
```

```python
# Bar Plot in seaborn


sns.barplot(data=tips, x='day',y='total_bill',palette = "viridis")
plt.title('Bar Plot of Total Bill')
plt.show()
```

```python
# Box plot in seaborn


sns.boxplot(data=tips, x='day',y='total_bill',palette = "coolwarm",hue='sex')
plt.title('Box Plot of Total Bill')
plt.show()
```

```python
# Scatter plot

sns.scatterplot(data=tips, x='total_bill',y='tip',hue='day',style='time',palette = 'deep')
plt.title('Scatter Plot of Total Bill and Tip')
plt.show()
```

```python
# line plot

sns.lineplot(data=tips, x='size',y='tip',hue='day',marker ='o',color = 'green')
plt.title('Line Plot of Total Bill and Tip')
plt.show()
```

```python
# pair plot

sns.pairplot(data=tips,hue='sex',palette = 'magma')
plt.show()
```

```python
# correlation heat map


numeric_tips = tips.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numeric_tips.corr()
sns.heatmap(correlation_matrix,annot=True,cmap='cubehelix',fmt='.2f')
plt.title('Correlation Heat Map')
plt.show()
```



># Task 1 :
- 1) How does passenger class(Pclass) relate to survival rates?
     * Use a bar plot to compare the survival rates across different passenger classes.
     * Explanation : A bar plot is effective for comparing categorical data. By plotting the mean survival rate for each passenger class, we can observe how class influenced the likelihood of survival.
     * Dataset : train.csv

```python
train = pd.read_csv('train.csv')
train
```

```python
sns.barplot(data=train, x='Pclass',y='Survived',palette = "rocket",errorbar=None)
plt.title('Bar Plot of Survival Rate')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```



> # Task 2 :
- Is there a relationship between passenger age and fare?
- Create a scatter plot to visualize the relationship between age and fare.
-Explanation : Scatter plots are ideal for examining the relationship between two continuous variables. This plot will help identify any patterns or correlations between a passengers age and the fare they paid
-Dataset : train.csv




```python
sns.scatterplot(data=train, x='Age',y='Fare',hue='Survived',palette = 'crest',s=100)
plt.title('Scatter Plot of Survival Rate')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```



> # Task 3 :
- What are the pairwise relationships between age,fare and passenger class?
- Use a pair plot to visualize the pairwise relationship among these variables.
- Explanation : Pair plots provide a matrix of scatter plots fro each pair of variables, along with histograms for individual variables . This visualization helps in nuderstanding the distribution and relationships between multiple variables simultaneously
- Dataset : train.csv


```python
subset = train[['Age','Fare','Pclass']]
sns.pairplot(data=subset,palette = 'rocket_r',diag_kind='kde',plot_kws={'s': 100})
plt.suptitle('Pair Plot of Age, Fare and Passenger Class')
plt.show()
```



> # Task 4
- What is the correlation between different numerical features in the dataset?
- Compute the correlation matrix and visualize it using a heatmap
- Explanation : A heatmap of the correlation matrix allows us to see the strength and direction of relationships between numerical variables. This can highlight which features are strongly correlated, aiding in feature selected and unedrstanding data dynamics
- Dataset : train.csv



```python
numerical_train = train.select_dtypes(include=['number'])
correlation_matrix = numerical_train.corr()
sns.heatmap(correlation_matrix,annot=True,cmap='Blues',fmt='.2f',linewidth=0.5)
plt.title('Correlation Heat Map')
plt.show()
```
