```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli
p=0.6
data = bernoulli.rvs(p,size=1000)
plt.bar([0,1],[sum(data==0),sum(data==1)],color=['red','green'],tick_label=['Failure(0)','Success(1)'])
plt.xlabel("Outcome")
plt.ylabel("Frequency")
plt.title("Bernoullis Distributin (p=0.6)")
plt.show()
```

```python
from scipy.stats import binom
n=10
p=0.3
x=np.arange(0,n+1)
pmf = binom.pmf(x,n,p)
plt.bar(x,pmf,color='blue',alpha=0.7)
plt.xlabel("Number of success (clicks)")
plt.ylabel("Probability")
plt.title(f"Binomial Distribution (n={n},p={p})")
plt.show()
```

```python
from scipy.stats import norm
mui = 70
sigma = 10
x=np.linspace(mui - 4 * sigma, mui + 4 * sigma,100)
pdf = norm.pdf(x, mui, sigma)
plt.plot(x,pdf,color="red", label=f'N({mui},{sigma})')
plt.xlabel("Exam score")
plt.ylabel("probability density")
plt.title(f"Normal Distribution ( ={mui}, = {sigma}))")
plt.legend()
plt.grid()
plt.show()
```

```python
from scipy.stats import norm
z=np.linspace(-4, 4, 1000)
pdf=norm.pdf(z,0,1)
plt.plot(z,pdf,color='yellow',label="Standard Normal Distribution(=0,=1)")
plt.xlabel("Z-Score")
plt.ylabel("Probability Density")
plt.title("Standard Normal Distribution (Z-Distribution)")
plt.axvline(x=0,color='red',linestyle='--',label="Mean(z=0)")
plt.legend()
plt.grid()
plt.show()
```

```python
import seaborn as sns
np.random.seed(42)
income_data = np.random.exponential (scale=50, size=1000)
plt.figure(figsize=(8, 4))
sns.histplot(income_data, bins=50, kde=True, color="r")
plt.title("Original Income Distribution (Skewed)")
plt.xlabel("Income ($1000)")
plt.ylabel("Frequency")
plt.show()
sample_means = []
n_samples= 1000
sample_size= 30
```

```python
for _ in range(n_samples):
    sample = np.random.choice(income_data, size=sample_size)
    sample_means.append(np.mean(sample))
plt.figure(figsize=(8, 4))
sns.histplot(sample_means, bins= 50, kde= True, color="b")
plt.title(f"Sampling Distribution of Sample Mean (n={sample_size})")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()
```
