```python
# import numpy as np
# x = np.array([1,2,3,4,5])
# y = np.array([3,4,2,4,5])
# def LR (x,y):
#   xm = np.mean(x)
#   ym = np.mean(y)
#   n= len(x)
#   num = 0
#   de = 0
#   for i in range(n):
#     num += (x[i] - xm) * (y[i] - ym)
#     de += (x[i] - xm)**2
#   mean = num/de
# mean

def LR(x,y):
  meanx = sum(x)/len(x)
  meany = sum(y)/len(y)

  mean = 0
  num =0
  de =0
  for i in range(len(x)):
    num += (x[i] - meanx) * (y[i] - meany)
    de += (x[i] - meanx)**2
  mean += num/de
  c = meany - mean*meanx
  return mean,c

def predict (x):
  y_pred =[]
  for i in range(len(x)):
    y_p = x[i]*mean + c
    y_pred.append(y_p)
  return y_pred

def r2(y,y_pred):
  numr =0
  der=0
  for i in range(len(y)):
    numr += (y_pred[i] - meany)**2
    der += (y[i] - meany)**2
  error_rate = numr/der
  return error_rate


x = [1,2,3,4,5]
y = [3,4,2,4,5]
mean,c = LR(x,y)
print(f"meany = {meany}")
print(f"y = {mean:.2f}x + {c:.2f}")
p = predict(x)
print(f"predicted values : {p}")
err = r2(y,p)
print(f"error rate : {err:.2f}")
accuracy = 1- err
print(f"accuracy : {accuracy:.2f}")
```
