```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# List of image file paths (replace with your actual image paths)
image_paths = ['image1.png', 'image2.png', 'image3.png', 'image4.png', 'image5.png',
               'image6.png', 'image7.png', 'image8.png', 'image9.png', 'image10.png',
               'image11.png', 'image12.png', 'image13.png', 'image14.png', 'image15.png']

# Create a figure with subplots (1 column, n rows where n is the number of images)
# Increase the figsize to make the images larger
fig, axes = plt.subplots(len(image_paths), 1, figsize=(10, len(image_paths)*6))

# Loop through each image path and display it in the corresponding subplot
for i, img_path in enumerate(image_paths):
    img = mpimg.imread(img_path)
    axes[i].imshow(img)
    axes[i].axis('off')  # Hide axis

# Show the plot
plt.show()
```
