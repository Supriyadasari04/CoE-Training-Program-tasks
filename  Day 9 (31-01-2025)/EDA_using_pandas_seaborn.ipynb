```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Corrected dataset
sd = {'Difficulty': ['Easy', 'Medium', 'Hard', 'Medium', 'Easy'],
      'TestScore': [80, 40, 20, 60, 70]}  # List of scores instead of a single string

q = pd.DataFrame(sd)

# Label encoding for the 'Difficulty' column
encoder = LabelEncoder()
q['Difficulty'] = encoder.fit_transform(q['Difficulty'])

# Print the modified DataFrame
print(q)

# Calculate the correlation
correlation = q['Difficulty'].corr(q['TestScore'])
print('Correlation:', correlation)
```

```python
data = {'colors' : ['RED','BLUE','GREEN','BLUE','RED'],
        'TestScore' : [50,60,70,80,90]}
df = pd.DataFrame(data)
df_encoded = pd.get_dummies (df,columns=['colors'])
print(df_encoded)
correlation = df_encoded['colors_RED'].corr(df_encoded['TestScore'])
print('Correlation with red:', correlation)
```
