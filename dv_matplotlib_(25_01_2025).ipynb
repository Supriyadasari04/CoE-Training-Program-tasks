```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```



> 2) Bar plot - which countries had the highest average temperatures in the dataset? Use case : compare categories or groups.


```python

country = pd.read_csv('GlobalLandTemperaturesByCountry.csv')

country.dropna(inplace=True)
country.isnull().sum()
```

```python
country_avg_temp = country.groupby('Country')['AverageTemperature'].mean().reset_index()
country_avg_temp
```

```python
country_avg_temp=country_avg_temp.dropna()
```

```python
top_countries = country_avg_temp.sort_values(by='AverageTemperature', ascending=False).head(10)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(top_countries['Country'], top_countries['AverageTemperature'], color='orange')
plt.xlabel('Country', fontsize=12)
plt.ylabel('Average Temperature (°C)', fontsize=12)
plt.title('Top 10 Countries with Highest Average Temperatures', fontsize=16)
plt.xticks(rotation=45)
```



> 3)  Histogram : what is the distrubution of average temperature across all data?




> use case : understand the frequency of values in  a dataset




> example plot : xaxis - temperature ranges, yaxis - frequency


```python
plt.figure(figsize=(10, 6))
plt.hist(country['AverageTemperature'], bins=10, color='purple', edgecolor='black')

plt.title("Distribution of Average Temperature", fontsize=16)
plt.xlabel("Temperature (°C)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)

plt.tight_layout()
plt.show()
```



> 4)  Scatter plot : Is there a correlation between temperature adn uncertainity in the dataset?

> Use Case : Explore relationship between two continuous variables

> Example Plot : xaxis - average temperature, yaxis - average temperature uncertainity







```python

plt.figure(figsize=(10, 6))
plt.scatter(country['AverageTemperature'], country['AverageTemperatureUncertainty'], color='lavender',  marker="o",s=100,edgecolor ='purple')

plt.title("Scatter Plot: Temperature vs. Uncertainty", fontsize=16)
plt.xlabel("Average Temperature (°C)", fontsize=12)
plt.ylabel("Average Temperature Uncertainty (°C)", fontsize=12)


plt.tight_layout()
plt.show()
```
