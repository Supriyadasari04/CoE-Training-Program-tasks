> Sales analysis : Use line plot to track monthly sales and histograms for order quantity distribution

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#monthy sales -line plot


months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales = [4500, 4800, 5000, 5200, 5500, 6000, 6200, 6500, 6800, 7000, 7200, 7500]
plt.plot(months, sales, marker='*', color='blue', linestyle='-', linewidth=2,markerfacecolor='red',markersize=10)
plt.title('Monthly Sales ')
plt.xlabel('Month')
plt.ylabel('Sales in Rs.')
plt.grid(True)
plt.show()


```

```python
#order quantity distribution -histogram

order_quant = [10, 15, 20, 25, 30, 35, 40, 20, 25, 30, 10, 50, 60, 70, 80, 90, 100]
plt.hist(order_quant, bins=10, color='crimson', edgecolor='black')
plt.title('Order Quantity Distribution')
plt.xlabel('Order Quantity')
plt.ylabel('Frequency')
plt.grid(False)
plt.show()
```

> Customer Demographics : Visualize age distribution (histogram) and spending patterns (scatter plot)

```python
#customer age distribution -histogram


ages = [22, 24, 29, 35, 40, 55, 60, 30, 23, 27, 22, 21, 25, 38, 40, 50, 46, 34, 60, 63]
plt.hist(ages, bins=10, color='cyan', edgecolor='black')
plt.title('customer age Distribution')
plt.xlabel('Customer Age')
plt.ylabel('Frequency')
plt.grid(False)
plt.show()

```

```python
#customer spending patterns -scatter plot

spending = [200, 250, 300, 350, 400, 450, 500, 550, 600, 650, 700, 750, 800, 850, 900, 950, 1000, 1050, 1100, 1150]
plt.scatter(ages, spending, color='purple', alpha=0.75)
plt.title('Customer Age vs Spending')
plt.xlabel('Age')
plt.ylabel('Spending Rs.')
plt.grid(True)
plt.show()
```

> School Performance : Compare students marks across subjects (bar chart) and identify outliers ( box plot)

```python
#school performance:compare students' marks across subjects(barchart)

subjects = ['Math', 'Science', 'English', 'Telugu']
marks = [92, 85, 76, 88]
plt.bar(subjects, marks, color='lightgreen')
plt.title('Students  Marks Across Subjects')
plt.xlabel('Subject')
plt.ylabel('Marks')
plt.show()
```

```python
#IDENTIFYING OUTLINERS

plt.boxplot(marks, vert=False)
plt.title('Box Plot for Marks Across Subjects')
plt.xlabel('Marks')
plt.show()
```
